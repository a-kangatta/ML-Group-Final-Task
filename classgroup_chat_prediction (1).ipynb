```python
import spacy
import numpy as np
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt
import nltk
import json
```

```python
nltk.download('punkt')
```

```python
from tensorflow.keras.layers import SimpleRNN, GRU, LSTM,Embedding, InputLayer, Input
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.tokenize import word_tokenize, sent_tokenize
from sklearn.metrics import accuracy_score
from sklearn.decomposition import PCA
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.optimizers import Adam
```

```python
data=open("/content/drive/MyDrive/Machine Learning/_chat.txt")
data=data.read()
data
```

```python
word_tokenize(data)
```

```python
#make a list
chat_list=[]
chats=sent_tokenize(data)
for sent in data:
  chat_list.append([sent])
```

```python
chats
```

```python
len(chat_list)
```

```python
vectorizer=TfidfVectorizer()
```

```python
x=vectorizer.fit_transform(chats)
```

```python
vocab=vectorizer.get_feature_names()
len(vocab)#, vocab 
```

```python
#ctreate embeddings matrix and transposing it
e_array=x.toarray().T
e_array
```

```python
#creating a dense embedding
#ie, each word will be represented by a vectors of 100 dimensions
pca=PCA(n_components=250)
```

```python
dense_embed=pca.fit_transform(e_array)
dense_embed
```

```python
#embeddings dictionary
embed_dict={}
for i in range(len(vocab)):
  embed_dict[vocab[i]]=dense_embed[i].tolist()
#21st word becomes the target
seq_len=20
tokenizer=Tokenizer()
tokenizer.fit_on_texts([data])
x=tokenizer.texts_to_sequences([data])
```

```python
len(x),len(x[0])
```

```python
#tokenizer.get_config
vocab_dict=json.loads(tokenizer.get_config().get('word_index'))
vocab_dict
```

```python
#keys and values
values=list(vocab_dict.values())
keys=list(vocab_dict.keys())
```

```python
vocab_len=len(keys)
vocab_len
```

```python
e_array=np.zeros((vocab_len+1,250))
e_array.shape
```

```python
for word,i in tokenizer.word_index.items():
  if word in embed_dict:
    embed_vector=embed_dict[word]
    e_array[i]=embed_vector #where the vector isnt none
```

```python
len(x[0])
```

```python
x[0][0:15]
```

```python
inner_list=[]
X={}
corp_len=len(x[0])-seq_len
k=0

for i in range(0, corp_len, seq_len):
  inner_list.clear()
  for j in range(i, i+seq_len):
    inner_list.append(x[0][j])
  X[k]=inner_list[:]
  k=k+1
```

```python
#list xvalues
x_train=[]
y_train=[]
for num in list(X.values()):
  x_train.append(num[0:-1])
  y_train.append(num[-1])
```

```python
len(x_train)
```

```python
X=np.array(x_train)
y=np.array(y_train).reshape(len(y_train),1)
```

```python
X.shape, y.shape
```

```python
y_train
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y, test_size=0.2, random_state=101)
```

```python
#x_train= input_seq
#x_hat=test_seq
#create tensors to put the matriz in
input_seq,test_seq,y_true,y_hat=tf.constant(x_train),tf.constant(x_test),tf.constant(y_train),tf.constant(y_test)
                                                         
```

```python
len(X), len(input_seq),len(test_seq)
```

```python
input_seq
```

```python
#prediction/ target
y_true
```

```python
#loss function is sparse_categorical_crossentropy, because the dataset is rather large
#the metrics beung tracked is accuracy as we want to see how accurately the model can predict the next word(s)
```

```python
#create the model
model_1=tf.keras.Sequential([
                             tf.keras.layers.Embedding(input_dim=vocab_len+1, output_dim=250, trainable=True, weights=[e_array]),
                             tf.keras.layers.LSTM(128, return_sequences=True),
                             tf.keras.layers.LSTM(128, return_sequences=True),
                             tf.keras.layers.LSTM(128, activation="relu"),
                             tf.keras.layers.Dense(100, activation="relu"),
                             tf.keras.layers.Dense(100, activation="relu"),
                             tf.keras.layers.Dense(vocab_len, activation="softmax")
                             ])
model_1.summary()
#compile model
model_1.compile(loss=tf.keras.losses.sparse_categorical_crossentropy,
                optimizer=tf.keras.optimizers.Adam(learning_rate=0.01),
                metrics=['accuracy']
                 )
#fit/train
model_1.fit(input_seq, y_true, epochs=130, verbose=1)
```

```python
#Improving the model
#1..Better embeddings ie incr the output dimensions
##ie at pca=PCA(n_components=100) 
#and 
##at e_array=np.zeros((vocab_len+1,200))

#2..train for longer
#3..more epochs
#4..more layers
```

```python
#evaluate model against test set
model_1.evaluate(test_seq, y_hat)
```
